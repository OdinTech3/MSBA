This goal of this notebook is to demonstrate the following:

* Read a file for data
* Produce summary statistics of the data

The data set we are working with is a csv file containing the years of experience and salary of employees in fictitious company called Happy Co.

Lets read the csv into a variable called `company_df`.

```r
company_df <- read.csv(
  file = "./data/salary_data.csv",
  header = T,
  sep = ",",
  na.strings = "?",
  stringsAsFactors = F
)
```

Let's get the first 5 rows of the dataset

```r
company_df[1:5,]
```

Below is a summary of the dataset that R produces. It is a summary about each column of data in the dataset.

```r
summary(company_df)
```

```r
sprintf("The total number of employees in Happy Co. is %d", nrow(company_df))
```

Using a scatter plot, we can show the correlation between years of experience and salary.

```r
plot(
  x = company_df$YearsExperience,
  y = company_df$Salary,
  xlab = "Experience",
  ylab = "Salary",
  main = "Salary vs Experience"
)
```
